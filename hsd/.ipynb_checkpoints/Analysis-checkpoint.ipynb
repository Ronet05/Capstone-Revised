```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import csv
import os
import glob
```

```python
os.chdir("/home/ronet/Desktop/strtsmrt-master/data/hsd")
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
joint_csv=pd.DataFrame()
for f in all_filenames:
    df=pd.read_csv(f)
    df['Stock_Name']=f[:-4]
    joint_csv=pd.concat([joint_csv,df])

joint_csv.to_csv("stocks_trend.csv", index=False, encoding='utf-8-sig')
```
